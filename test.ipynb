```python
lab_l_channel = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)[..., 0]
```

```python
tsize = 240
```

```python
tiles = view_as_windows(lab_l_channel, (tsize, tsize), (tsize, tsize))
```

```python
lightness = tiles.reshape(-1, tsize ** 2).mean(axis=-1).reshape(np.array(img.shape[:2]) // tsize)
```
